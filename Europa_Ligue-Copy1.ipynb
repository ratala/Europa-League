```python
pages = []
joueur = []
club = []
nation = []
valeur = []
age = []

#https://www.transfermarkt.co.uk/europa-league/marktwerte/pokalwettbewerb/EL/ajax/yw1/gruppe//pos//detailpos//altersklasse/u23/plus/1/galerie/0/page/7

headers = {'User-Agent': 
           'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'}

for i in range(1, 11):
    url = "https://www.transfermarkt.co.uk/europa-league/marktwerte/pokalwettbewerb/EL/ajax/yw1/gruppe//pos//detailpos//altersklasse/u23/plus/1/galerie/0/page/{}".format(i)
    pages.append(url)

for item in pages:
    pageTree = requests.get(item, headers=headers)
    pageSoup = BeautifulSoup(pageTree.text, 'html.parser')
    for table in pageSoup.find_all('table', {'class': "items"}):  
        for Players in table.find_all("a", {"class": "spielprofil_tooltip"}):
            joueur.append(Players.text)
            
        for td in pageSoup.find_all("td", {"class": "rechts hauptlink"}):
            de = td.get_text()
            valeur.append(de)

        for tr in pageSoup.find_all("tr", {"class": ["even", "odd"]}):
            for td in tr.find_all("td", {"class": 'zentriert'},recursive = False):
                for a in td.find_all("a", {"class": 'vereinprofil_tooltip'},recursive = False):
                    img = a.find('img',recursive = False)
                    de = img['alt']
                    club.append(de)
                    
        
        for td in pageSoup.find_all("td", {"class": "zentriert"}):
            for img in td.find_all('img', {'class': "flaggenrahmen"}, limit = 1):
                dd = img['alt']
                nation.append(dd)

        for tr in pageSoup.find_all("tr", {"class": ["even", "odd"]}):
            for td in tr.find_all('td', "zentriert")[2]:           
        #te = td.get_text()  
                age.append(td)

   
##for tr in pageSoup.find_all("tr", {"class": "even"}): Continuer ici!
  #      for td in pageSoup.find_all("td", {"class": "rechts hauptlink"}):
  #          df = td.get_text()
    #        values.append(df)


            
    
club[249]
#pageSoup
```
